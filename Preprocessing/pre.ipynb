```python
def seperateByBar(s,d):
    note = s[0]
    beat = s[1]
    out = []
    tmp = []
    current_beat = 0
    last_note = ''
    for i in range(len(note)):
        if(beat[i]//d == current_beat):
            if(len(tmp)==0):
                tmp.append(last_note)
            tmp.append(note[i])
        else:
            out.append(tmp)
            tmp = []
            last_note = note[i]
            while(current_beat  < (beat[i]//d) - 1):
                out.append([])
                current_beat+=1
            current_beat= beat[i]//d
    return out
```

```python
for i in [4,12,36,48]:
    df['right_sep_' + str(i)] = df['right'].apply(lambda x: seperateByBar(x,i))
    df['left_sep_' + str(i)] = df['left'].apply(lambda x:seperateByBar(x,i))
```

```python
df['right_sep'] = df['right'].apply(lambda x: seperateByBar(x,12))
df['left_sep'] = df['left'].apply(lambda x:seperateByBar(x,12))
```
